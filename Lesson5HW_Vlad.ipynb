1. Напишите функцию generate_password(length), которая генерирует случайный пароль заданной длины, содержащий:

заглавные и строчные латинские буквы,

цифры,

символы !@#$%^&*.

Функция должна гарантированно включать хотя бы один символ из каждой категории.

```python
import random as rnd
import string

alphabet = string.ascii_letters

def gen_pass(length):
    new_pass = ''

    for i in range(length):
        rnd_let = rnd.choice(alphabet)
        rnd_num = rnd.randint(0, 9)
        rnd_symb = rnd.choice("!@#$^&*.,")
        new_pass += rnd_let + str(rnd_num) + rnd_symb
    return new_pass[0:length]

gen_pass(10)
```

2.
Напишите программу, которая запрашивает у пользователя текст и выводит:

количество слов,

количество уникальных слов,

самое часто встречающееся слово (и сколько раз),

среднюю длину слова.

```python
from collections import Counter as cou

stri = "Количество слов, количество уникальных слов, количество самого часто встречающеесего слова, и сколько раз, среднюю длину слова"

def word_average(stri):
    split_stri = stri.lower().split()
    len_spl_st = len(split_stri)
    set_len_spl_st = len(set(split_stri))
    most_use = cou(split_stri).most_common(1)
    len_word = 0
    for i in split_stri:
        len_word += len(i)
    num_word = len_word / len_spl_st
    return len_spl_st, set_len_spl_st, most_use, num_word

print(word_average(stri))

```

3.
Компьютер загадывает число от 1 до 100. Пользователь пытается угадать. После каждой попытки компьютер говорит «больше», «меньше» или «угадал». В конце вывести количество попыток.

```python
import random

def tree_of_choice():
    num2 = random.randint(0, 100)
    print(num2)
    num1 = int(input('Введите число'))
    while num1 != num2:
        if num1 > num2:
            print('Больше!')
            num1 = int(input())
        elif num1 < num2:
            print('Меньше!')
            num1 = int(input())
    print('Угадал!')

tree_of_choice()
```

4.
Напишите программу «Таймер». Пользователь вводит количество секунд. Программа отсчитывает время, каждую секунду выводя оставшееся время в формате MM:SS. По истечении времени выводится сообщение и проигрывается звуковой сигнал (print('\a')).

```python
import time
import datetime

def timer():
    seconds = int(input('Введите количество секунд'))

    while seconds > 0:
        user_time = datetime.time(0, seconds//60, seconds % 60)
        form_time = user_time.strftime('%M:%S')

        print(form_time + "     ", end='\r')
        time.sleep(1)
        seconds -= 1

    return "\aЗвуковой сигнал"

print(timer())
```
