1. Напишите функцию для разделения строки и преобразования её в список слов. Например "Robin Singh" ==> ["Robin", "Singh"]

```python
input_string = "Robin Hood Singh Song"
result_list = []

def separation_string(input_string):
    new_string = ""
    for letter in input_string:
        new_string += letter
        if " " in letter:
            new_string = new_string.replace(" ", "")
            result_list.append(new_string)
            new_string = ""
    result_list.append(new_string)
    return result_list

print(separation_string(input_string))
```

2. Натан любит кататься на велосипеде. Поскольку Натан знает, как важно избегать обезвоживания, он выпивает 0,5 литра воды за час езды на велосипеде. Вам дано время в часах, и вам нужно вернуть количество литров, которое выпьет Натан, округленное до наименьшего значения. Пример time = 3 ----> litres = 1

```python
from datetime import datetime

def lit_per_hour():
    time_in = datetime.strptime((input("Введите количество часов и минут в пути в виде 00:00")), "%H:%M")
    int_hours = time_in.hour
    result = (int_hours*0.5)//1
    return result

print(lit_per_hour())


```

3. Создайте функцию, которая возвращает список целых чисел от n до 1, где n>0. Пример: n=5-->[5,4,3,2,1]

```python
some_list = [5, 5.2, 6, 2, 1.5, 3.33333, 7, 8, 1, 7.7]
int_list = []

def integer_list(some_list):
    for num in some_list:
        if (num//1) == num:
            int_list.append(num)
    return len(int_list)

print(integer_list(some_list))
```

4. Учитывая год, верните столетие, в котором он находится. Например 1705 --> 18    1900 --> 19     1601 --> 17

```python
some_year = int(input("Введите дату, например 1990"))

if some_year > 100:
    result = some_year//100 + 1
elif some_year < 100 and some_year >= 0:
    result = 1

print(result)
```

5. Учитывая случайное неотрицательное число, вам необходимо вернуть цифры этого числа в массиве в обратном порядке. 35231 => [1,3,2,5,3]

```python
some_num = str(35231)
result_list = []

for i in range(len(some_num) - 1, -1, -1):
    result_list.append(some_num[i])

print(result_list)
```
